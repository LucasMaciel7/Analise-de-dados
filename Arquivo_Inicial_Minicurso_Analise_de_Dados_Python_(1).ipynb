````python
import pandas as pd

tabela = pd.read_csv("/content/drive/MyDrive/Mini curso analise de dados /ClientesBanco (1).csv", encoding="latin1")
tabela=tabela.drop("CLIENTNUM",axis=1)
tabela=tabela.drop("Sexo",axis=1)
````

Eliminando Valores Nulos


````python

tabela=tabela.dropna()
display(tabela.describe().round(1))
````

Vendo o porcentual de clientes cancelados

````python
qtd_categoria= tabela["Categoria"].value_counts(normalize="true")
print(qtd_categoria)
````

Fazendo a comparação entre o cliente cancelado com o restante das colunas, para ver se achamos algum ponto em comum

````python
import plotly.express as px

for coluna in tabela:
  grafico= px.histogram(tabela,x=coluna, color="Categoria")
  display(grafico)
````
